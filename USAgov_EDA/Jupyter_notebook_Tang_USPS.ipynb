### Data Cleansing 

```python
import re 

def extract_file_name_from_url(my_str):
    new_str = re.sub('\s\|.*', '', my_str)
    new_str = re.sub('-paper-version', '', new_str)
    new_str = re.sub('PDF', 'pdf', new_str)
    new_str = re.sub('Welcome', 'July 2002 Price List', new_str)
    new_str = re.sub('U.S. Postal Service Announces New Prices for 2022 - Newsroom - About.usps.com', 'July 2002 Price List', new_str)
    new_str = re.sub('Retail Postage Price Calculator Mail Services', 'July 2002 Price List', new_str)
    new_str = re.sub('Retail Postage Price Calculator', 'July 2002 Price List', new_str)
    new_str = re.sub('Manuals', 'Employee Relations', new_str)
    new_str = re.sub('- Targeted Mail Marketing', 'User Guide', new_str)
    new_str = re.sub('PO Boxes', 'Customer Agreement for Premium PO Box Service', new_str)
    return new_str
```

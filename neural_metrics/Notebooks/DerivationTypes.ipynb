```python
# Non-positional features

def absent_terms(candidate, reference):
  output = 0
  for token in np.unique(reference):
    if token not in candidate:
      output += 1

  return output / len(np.unique(reference))

def novel_terms(candidate, reference):
  output = 0
  for token in np.unique(candidate):
    if token not in reference:
      output += 1

  return output / len(np.unique(candidate))

def length_ratio(candidate, reference):

  return len(candidate) / len(reference)

# Engrams

def engrams(tokens, n):
  output = []
  for i in range(len(tokens) - n + 1):
    output.append([tokens[i]])
    for j in range(1, n):
      output[-1].append(tokens[i + j])

  return output

def engram_comparison(candidate, reference, n):
  c_engrams = engrams(candidate, n)
  r_engrams = engrams(reference, n)

  if len(r_engrams) == 0 or len(c_engrams) == 0:
    return 0

  output = 0
  for re in r_engrams:
    if re in c_engrams:
      output += 1

  return output / len(r_engrams)

# Positional features

# TODO Deletion

# TODO Mutation

# TODO Insertion
```

```python
def mutation_types(candidate, reference, smallest_seen_engram = np.inf):
  if len(candidate) == 0 and len(reference) == 0:
    return []
  if len(candidate) == 0 and len(reference) != 0:
    return ["DELETION-" + str(len(reference))]
  if len(candidate) != 0 and len(reference) == 0:
    return ["INSERTION-" + str(len(candidate))]
  if candidate == reference:
    return candidate
  for engram_length in range(min(min(len(candidate), len(reference)), smallest_seen_engram), 0, -1):
    c_engrams = engrams(candidate, engram_length)
    r_engrams = engrams(reference, engram_length)
    for i in range(len(r_engrams)):
      for j in range(len(c_engrams)):
        if r_engrams[i] == c_engrams[j]:
          return mutation_types(candidate[:j], reference[:i], engram_length) + c_engrams[j] + mutation_types(candidate[j + engram_length:], reference[i + engram_length:], engram_length)

  return ["SUBSTITUTION-" + str(max(len(candidate), len(reference)))]
```

```python
df = df.loc[df['lp'] == 'en-de'].dropna()
```

```python
ref_token_count = []
cand_token_count = []
mutation_strings = []
ats = []
nts = []
lr = []
engrams_1 = []
engrams_2 = []
engrams_3 = []
engrams_4 = []
engrams_5 = []
substitution_counts = []
deletion_counts = []
insertion_counts = []

for index, row in df.iterrows():
  if len(ref_token_count) % 100 == 0:
    print(len(ref_token_count) / len(df['mt']))
  candidate = tokenizer.tokenize(row['mt'])
  reference = tokenizer.tokenize(row['ref'])
  ref_token_count.append(len(reference))
  cand_token_count.append(len(candidate))
  ats.append(absent_terms(candidate, reference))
  nts.append(novel_terms(candidate, reference))
  lr.append(length_ratio(candidate, reference))
  engrams_1.append(engram_comparison(candidate, reference, 1))
  engrams_2.append(engram_comparison(candidate, reference, 2))
  engrams_3.append(engram_comparison(candidate, reference, 3))
  engrams_4.append(engram_comparison(candidate, reference, 4))
  engrams_5.append(engram_comparison(candidate, reference, 5))
  mutation_strings.append(mutation_types(candidate, reference))
  substitutions = []
  deletions = []
  insertions = []
  for item in mutation_strings[-1]:
    if "SUBSTITUTION" in item:
      substitutions.append(int(item.split("-")[-1]))
    if "DELETION" in item:
      deletions.append(int(item.split("-")[-1]))
    if "INSERTION" in item:
      insertions.append(int(item.split("-")[-1]))
  insertion_counts.append(insertions)
  deletion_counts.append(deletions)
  substitution_counts.append(substitutions)
```

```python
df['Reference Token Count'] = ref_token_count
```

```python
df['Candidate Token Count'] = cand_token_count
df['Mutation String'] = mutation_strings
df['Absent Terms'] = ats
df['Novel Terms'] = nts
df['Length Ratio'] = lr
df['Engram Overlap 1'] = engrams_1
df['Engram Overlap 2'] = engrams_2
df['Engram Overlap 3'] = engrams_3
df['Engram Overlap 4'] = engrams_4
df['Engram Overlap 5'] = engrams_5
df['Substitutions'] = substitution_counts
df['Deletions'] = deletion_counts
df['Insertions'] = insertion_counts
```
